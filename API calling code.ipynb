```python
import csv
import os
import openai
import csv
import json
```

```python

openai.api_key = 'your key here'

def generate_auto_grader_response(essay_text):
    rubrics = '''
    Rubric for assessing language proficiency of English Language Learners (ELLs):

    Category: Cohesion

    Score Description
    0 Lack of organization and coherence; weak connections between ideas.
    1 Limited use of cohesive devices; inconsistent organization.
    2 Basic use of some cohesive devices; somewhat clear connections between ideas.
    3 Adequate use of cohesive devices; generally clear organization and connections.
    4 Effective use of cohesive devices; well-organized and connected ideas.
    5 Exceptional use of a wide range of cohesive devices; seamless and logical flow of ideas.

    Category: Syntax

    Score Description
    0 Frequent and significant errors in sentence structure; lack of variety.
    1 Limited sentence structure variety; noticeable errors.
    2 Basic variety in sentence structure with some errors.
    3 Adequate variety in sentence structure with occasional errors.
    4 Good variety in sentence structure with minor errors.
    5 Excellent variety in sentence structure with accurate and complex sentences.

    Category: Vocabulary

    Score Description
    0 Limited and repetitive vocabulary; frequent use of inappropriate words.
    1 Basic vocabulary with limited range and some inaccuracies.
    2 Adequate vocabulary range with occasional inaccuracies.
    3 Satisfactory vocabulary range with occasional advanced words.
    4 Good vocabulary range with few inaccuracies and appropriate word choice.
    5 Wide range of vocabulary, including advanced expressions; precise word choice.

    Category: Phraseology

    Score Description
    0 Frequent use of incorrect idioms or inappropriate expressions.
    1 Limited use of idiomatic expressions or inconsistent use of appropriate phrases.
    2 Basic use of some idiomatic expressions or inconsistent use of appropriate phrases.
    3 Adequate use of idiomatic expressions and some appropriate phrases.
    4 Good use of idiomatic expressions and appropriate phrases.
    5 Excellent use of idiomatic expressions and varied, appropriate phrases.

    Category: Grammar

    Score Description
    0 Numerous and significant grammar errors; lack of understanding of basic rules.
    1 Frequent grammar errors; limited control of basic rules.
    2 Some grammar errors; basic control of rules.
    3 Occasional grammar errors; moderate control of rules.
    4 Few grammar errors; good control of rules.
    5 Virtually no grammar errors; excellent command of rules.

    Category: Conventions

    Score Description
    0 Frequent and significant errors in punctuation, capitalization, and spelling.
    1 Limited control of punctuation, capitalization, and spelling; noticeable errors.
    2 Basic control of punctuation, capitalization, and spelling with occasional errors.
    3 Adequate control of punctuation, capitalization, and spelling with occasional errors.
    4 Good control of punctuation, capitalization, and spelling with minor errors.
    5 Excellent control of punctuation, capitalization, and spelling with near-perfect accuracy.
    '''

    prompt = f"Grade the following essay based on the provided rubrics:\n\nEssay text: \"{essay_text}\"\n{rubrics}"

    response = openai.Completion.create(
        model="text-davinci-003",
        prompt=prompt,
        temperature=0.2,
        max_tokens=60,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0
    )

    auto_grader = response.choices[0].text.strip()
    return auto_grader


def read_essay_text_from_csv(csv_file, num_rows=50):
    essay_text = []
    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        next(reader) 
        for i, row in enumerate(reader):
            essay_text.append(row[1]) 
            if i + 1 == num_rows:
                break
    return essay_text


csv_file_path = 'train.csv'
essay_text_list = read_essay_text_from_csv(csv_file_path, num_rows=50)

output_data = [] 

for essay_text in essay_text_list:
    auto_grader_response = generate_auto_grader_response(essay_text)
    output_data.append(auto_grader_response)


output_file_path = 'auto_grader_responses_1.json'
with open(output_file_path, 'w') as json_file:
    json.dump(output_data, json_file, indent=4)

print(f"Auto grader responses have been stored in {output_file_path}")
```

```python

with open('auto_grader_responses_1.json', 'r') as json_file:
    data = json.load(json_file)


split_data = []

for item in data:
    item_dict = {}
    values = item.split("\n")
    for value in values:
        if ": " in value:
            key, val = value.split(": ", 1)
            item_dict[key.strip()] = val.strip()
    split_data.append(item_dict)


for item in split_data:
    item.pop("Overall Score", None)
    item.pop("Overall Grade", None)


with open('output_10test.json', 'w') as json_file:
    json.dump(split_data, json_file, indent=4)
```

```python
with open('output_10test.json', 'r') as json_file:
    data = json.load(json_file)

fieldnames = data[0].keys()

with open('output_10test.csv', 'w', newline='') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data)
```
